```python
import pandas as pd
import numpy as np
import random
```

```python
# reading and showing dataframe
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

### Create Index

```python
# create unique values for the index column
new_index = np.arange(0,1000)
new_index
```

```python
# shuffle the index array
random.shuffle(new_index)
new_index
```

```python
# create a column with new_index
df_exams['new index'] = new_index
df_exams
```

```python
# set new_index column as index
df_exams.set_index('new index', inplace=True)
df_exams
```

```python
# sort the Data Frame by the index column
df_exams.sort_index()
```
