```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

## Filter a Data Frame based on One Condition 

### Check Company == Apple

```python
# find which rows have "Apple" in the "Company" column
df_laptops['Company']=='Apple'
```

```python
# filter dataframe based on condition
apple_laptops = df_laptops[df_laptops['Company']=='Apple']
len(apple_laptops)
```

```python
apple_laptops.head(3)
```

```python
# verify which companies are in the "Company" column with value_counts()
df_laptops[df_laptops['Company']=='Apple'].value_counts('Company')
```

### Check Company != HP

```python
# find which rows don't have "HP" in the "Company" column
df_laptops['Company']!='HP'
```

```python
# filter dataframe based on condition
not_hp_laptops = df_laptops[df_laptops['Company']!='HP'].value_counts('Company')
not_hp_laptops
```

### Check Price > 2000 euros

```python
# find laptops with price over 2000 euros
df_laptops[df_laptops['Price_euros']>2000]
```

## Filter a Data Frame based on Multiple Conditions 

### Apple laptops which costs more than 2000 euros - Using &(and) 

```python
apple_expensive = (df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] > 2000)
df_laptops[apple_expensive]
```

### Apple or Dell laptops - Using |(or) 

```python
apple_or_dell = (df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')
df_laptops[apple_or_dell]
```

### Apple or Dell laptops which cost more than a 2000 euros - Using | and & 

```python
apple_dell_expensive = ((df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')) & (df_laptops['Price_euros'] > 2000)
df_laptops[apple_dell_expensive]
```
