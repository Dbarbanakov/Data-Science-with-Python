### Filter 

```python
# find which rows don't have "HP" in the "Company" column
no_hp = df['Company'] != 'HP'
# filter dataframe based on condition
df[no_hp].head()
```

```python
# find laptops with price over 2000 euros
expensive = df['Price_euros'] > 2000
# filter dataframe based on condition
df[expensive].head()
```

### Conditional Column 

#### 2 Choices - np.where()

```python
# create an array based on screen size (screen size greather than 15 is 'Big' and less or equal to 15 is 'Small')
screen_size = np.where(df['Inches'] > 15, 'Big', 'Small')
screen_size
```

```python
# set it to a new column named 'Screen_size'
df['Screen_size'] = screen_size
# show first 5 rows
df.head()
```

```python
# count values in 'Screen_size' column
df['Screen_size'].value_counts()
```

#### More than 2 Choices - np.select()

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small<12

# create "conditions" and "values" variables
choices = ['Too Big', 'Big', 'Small', 'Too Small']
conditions = [df['Inches'] > 16,
             (df['Inches'] <= 16) & (df['Inches'] > 14),
             (df['Inches'] <= 14) & (df['Inches'] > 12),
             df['Inches'] <= 12,
             ]
```

```python
# set it to a new column "Screen_size"
df['Screen_size'] = np.select(conditions, choices)
```

```python
# show dataframe
df.head()
```

```python
# count values inside "Screen_size" column
df['Screen_size'].value_counts()
```
